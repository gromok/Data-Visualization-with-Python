### Load Gas Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5)) #changing the size
plt.title('Gass Prices over Time(in USD)', fontdict={'fontweight':'bold', 'fontsize':18}) #title of the gragh

#ploting values
#plt.plot(gas.Year,gas.USA,'b.-', label='USA')
#plt.plot(gas.Year,gas.Canada,'r.-', label='Canada')
#plt.plot(gas.Year,gas['South Korea'], 'g.-', label='South K')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::3])#solving x-axis issue

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gass_price_figure.png', dp1=300)
plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

### Histogram

```python
bins=[40,50,60,70,80,90,100]
plt.title('Distribution of Players Skill in Fifa 2018', fontdict={'fontweight':'bold', 'fontsize':18})
plt.hist(fifa.Overall, bins=bins, color='#03fccf')
plt.xticks(bins)

plt.xlabel('Skill Lavel')
plt.ylabel('No of Players')
plt.savefig('FIFA_Players_skills.png', dp1=300)

plt.show()
```

### Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
plt.title('Foot Refference for FIFA Players', fontdict={'fontweight':'bold', 'fontsize':18})
labels = ['Left', 'Right']
colors = ['#24a695', '#275d96']
explode=(left)

plt.pie([left, right], colors=colors, labels=labels, autopct='%.2f %%' )
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)== str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium =fifa.loc[(fifa.Weight >= 125) & (fifa.Weight <= 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight <= 175)].count()[0]
medium_heavy =fifa.loc[(fifa.Weight >= 175) & (fifa.Weight <= 200)].count()[0]
heavy = fifa.loc[fifa.Weight > 200].count()[0]

plt.title('Fifa Players Weight(in lbs)', fontdict={'fontweight':'bold', 'fontsize':18})
labels = ['0-125','125-150','150-175', '175-200', 'Above 200']
explode = (0.4, 0.1, 0,0, 0.6 )
plt.pie([light, light_medium, medium, medium_heavy, heavy], labels = labels, explode = explode, autopct='%.2f %%', shadow=True, startangle=180)

#plt.savefig('FIFA_Players_weight.png', dp1=300)
plt.show()
```

### Box and Whiskers

```python
plt.style.use('default')
plt.figure(figsize=(5,7))

barc = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
real = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

plt.title('Prefessional FIFA Clubs Comparison')

labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']
plt.boxplot([barc, real, revs], labels = labels)

plt.ylabel('Overall Rating')

#plt.savefig('FIFA_Club_Comparison.png', dp1=300)
plt.show()
```
